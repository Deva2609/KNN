```python
## Here Gender in string format so we have to convert string to float value.So replacing male->1 female->0
data['Gender'].replace('Male','1', inplace=True)
data['Gender'].replace('Female','0', inplace=True)
```

```python
X=data[['Gender','Age','Salary']]
y=data['Purchase Iphone']
```

train,test split method

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25)
```

```python
X_train.shape
```

```python
X_test.shape
```

Using KNN method

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
k=5
knn=KNeighborsClassifier(n_neighbors=k)
knn.fit(X_train,y_train)
```

```python
## Use standardization method
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
```

```python
StSc_X = StandardScaler()
X_train = StSc_X.fit_transform(X_train)
X_test = StSc_X.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
error_rate = []
for i in range(1,20):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
```

```python
for i in range(1,20):
    print("K = ",i,"Errors = ",error_rate)
```

```python
print("Minimun error :-", min(error_rate), "at K = ",error_rate.index(min(error_rate)))
```

```python
acc = []
for i in range(1,20):
    neigh = KNeighborsClassifier(n_neighbors=i).fit(X_train,y_train)
    yhat = neigh.predict(X_test)
    acc.append(accuracy_score(y_test,yhat))
```

```python
acc ## Accuracy is high when k=5
```

```python
k = 5
knn = KNeighborsClassifier(n_neighbors=k)
knn.fit(X_train,y_train)   #### 75% of the data is trained 
```
